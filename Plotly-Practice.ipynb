```python
import pandas as pd 
import numpy as np 
import plotly.express as px
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df=pd.read_csv("AirPassengers.csv",parse_dates=["Month"],index_col="Month")
df
```

```python
df.rename(columns={"#Passengers":"Passengers"})
```

```python
df.shape
```

```python
import plotly.express as px
fig = px.line(
    df,
    x=df.index,
    y="Passengers",
    labels={"x": "Month", "Passengers": "Passengers"},
    title="Air Passenger Trend"
)

fig.update_layout(
    width=1000,
    height=300,   # equivalent to figsize=(14,2)
    template="plotly_white"
)

fig.show()
```
