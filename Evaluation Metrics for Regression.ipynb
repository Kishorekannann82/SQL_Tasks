```python
y=df_train["SalePrice"]
```

```python
df_train=df_train.drop(["SalePrice"],axis=1)
combined=pd.concat([df_train,df_test],axis=0)
```

```python
combined.columns
```

```python
combined.shape
```

```python
num_cols=combined.select_dtypes(include=["float64","int64"]).columns
print(num_cols)
```

```python
combined[num_cols]=combined[num_cols].fillna(combined[num_cols].median())
```

```python
cat_cols=combined.select_dtypes(include=["object"]).columns
print(cat_cols)
```

```python
combined[cat_cols]=combined[cat_cols].fillna("None")
```

```python
combined=pd.get_dummies(combined,drop_first=True)
print(combined.shape)
```

```python
X_train=combined[:len(y)]
X_test=combined[len(y):]
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_tr,X_val,y_tr,y_val=train_test_split(X_train,y,test_size=0.2,random_state=42)
```

```python
model=LinearRegression()
model.fit(X_tr,y_tr)
y_pred=model.predict(X_val)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(y_val,y_pred)
mse=mean_squared_error(y_val,y_pred)
r2=r2_score(y_val,y_pred)
rmse=np.sqrt(mse)
print("Mean Absolute Error:",mae)
print("Mean Squareed Error:",mse)
print("Root Mean Squared Error:",rmse)
print("R2_Score:",r2)
```

```python
y=np.log1p(y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(X_train,y)
importances=rf.feature_importances_
feat_imp=pd.DataFrame({
    "feature":X_train.columns,
    "importance":importances}).sort_values(by="importance",ascending=False)
top_features=feat_imp.head(50)["feature"]
X_train_selected=X_train[top_features]
X_test_selected=X_test[top_features]
print(len(top_features))
```

```python
X_tr,X_val,y_tr,y_val=train_test_split(X_train_selected,y,test_size=0.2,random_state=42)
model=RandomForestRegressor(
    n_estimators=200,
    max_depth=10,
    random_state=42)
model.fit(X_tr,y_tr)
y_pred=model.predict(X_val)
y_pred=np.expm1(y_pred)
y_val_actual=np.expm1(y_val)
mae=mean_absolute_error(y_val_actual,y_pred)
rmse=np.sqrt(mean_squared_error(y_val_actual,y_pred))
r2=r2_score(y_val_actual,y_pred)
print("Improved MAE:",mae)
print("Improved RMSE:",rmse)
print("Improved R2:",r2)
```

```python
from xgboost import XGBRegressor
xgb_model=XGBRegressor(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=5,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42)
xgb_model.fit(X_tr,y_tr)
y_pred_xgb=xgb_model.predict(X_val)
y_pred_xgb=np.expm1(y_pred_xgb)
y_val_actual=np.expm1(y_val)
```

```python
def evaluate_metrics(y_true,y_pred,n_features):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    n=len(y_true)
    adj_r2=1 - (1 - r2) * (n - 1) / (n - n_features - 1)
    return mae,mse,rmse,r2,adj_r2
```

```python
rf_pred=model.predict(X_val)
rf_pred=np.expm1(rf_pred)
rf_metrics=evaluate_metrics(y_val_actual,rf_pred,X_val.shape[1])
```

```python
xgb_metrics = evaluate_metrics(y_val_actual, y_pred_xgb, X_val.shape[1])
```

```python
results = pd.DataFrame({
    "Model": ["Random Forest", "XGBoost"],
    "MAE": [rf_metrics[0], xgb_metrics[0]],
    "MSE": [rf_metrics[1], xgb_metrics[1]],
    "RMSE": [rf_metrics[2], xgb_metrics[2]],
    "R2": [rf_metrics[3], xgb_metrics[3]],
    "Adj_R2": [rf_metrics[4], xgb_metrics[4]]
})

print(results)
```
