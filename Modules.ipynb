## Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df4=sns.load_dataset("iris")
```

```python
df4.head()
```

```python
plt.plot(df4.index,df4["sepal_length"])
plt.title("Line Plot")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
df4.groupby("species")["sepal_length"].mean().plot(kind="bar")
plt.title("Avg Sepal Length")
plt.show()
```

```python
plt.hist(df4["sepal_length"])
plt.title("Histogram")
plt.show()
```

```python
plt.scatter(df4["sepal_length"],df4["petal_length"])
plt.title("Scatter Plot")
plt.xlabel("Sepal Length")
plt.ylabel("petal Length")
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10,8))

ax[0,0].plot(df4.index, df4["sepal_length"])
ax[0,0].set_title("Line")

ax[0,1].hist(df4["sepal_length"])
ax[0,1].set_title("Histogram")

ax[1,0].scatter(df4["sepal_length"], df4["petal_length"])
ax[1,0].set_title("Scatter")

df4.groupby("species")["sepal_length"].mean().plot(kind="bar", ax=ax[1,1])
ax[1,1].set_title("Bar")

plt.tight_layout()
plt.show()
```

## Plotly

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df5 = pd.DataFrame({
    'Month': ['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales': [100, 130, 120, 160, 150, 180],
    'Profit': [20, 30, 25, 40, 35, 50]
})
```

```python
df5
```

```python
fig=px.line(df5,x="Month",y="Sales",title="Monthly Sales")
fig.show()
```

```python
##Multiple Lines
fig=px.line(df5,x="Month",y=["Sales","Profit"],title="Sales vs Profit",markers=True)
fig.show()
```

```python
fig=px.bar(df5,x="Month",y="Sales",color="Sales",title="Monthly Sales",text="Sales")
fig.show()
```

```python
fig=px.bar(df4,x="species",y="sepal_length",title="Bar_Chart",text="sepal_length")
fig.show()
```

```python
fig=px.scatter(
    df4,
    x="sepal_length",
    y="petal_length",
    color="species",
    size="sepal_width",
    title="Scatter_Plot")
fig.show()
```

```python
count=df4["species"].value_counts().reset_index()
count.columns=["species","count"]
fig=px.pie(
    count,
    names="species",
    values="count",
    title="Pie_Chart")
fig.show()
```

```python
## 3d Plots
fig=px.scatter_3d(
    df4,
    x="sepal_length",
    y="sepal_width",
    z="petal_length",
    color="species",
    title="3D Plot")
fig.show()
```

```python
## 3d Plots
fig=px.line_3d(
    df4,
    x="sepal_length",
    y="sepal_width",
    z="petal_length",
    color="species",
    title="3D Plot")
fig.show()
```

```python
fig = px.line(
    df4,
    x=df4.index,
    y="sepal_length",
    title="Customized Plot"
)

fig.update_layout(
    xaxis_title="Index",
    yaxis_title="Sepal Length",
    template="plotly_dark"
)

fig.show()
```
